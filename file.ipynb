```python
#read specific columns from a csv file
import csv
name=input("enter the csv file name:")
with open(name,"r") as f1:
    reader=csv.DictReader(f1)
    print("\n Department Details:")
    for row in reader:
        print("DeptId:",row['DeptId'],"TotalEmployees:",row['TotalEmployees'])
        
```

```python
#write apython dictionary to a csv file then read it back
import csv

data=[
    {"Name":"sahva","Age":22,"Marks":75},
    {"Name":"sinu","Age":21,"Marks":79},
    {"Name":"aswani","Age":20,"Marks":70},
    {"Name":"nafla","Age":21,"Marks":75},
    {"Name":"saniya","Age":22,"Marks":75},
    
]
with open("dict.csv","w",newline='') as file:
          fieldnames=["Name","Age","Marks"]
          writer=csv.DictWriter(file,fieldnames=fieldnames)
          writer.writeheader()
          writer.writerows(data)
print("Dictionary data written successfully!") 
          
with open("dict.csv","r") as file:
          reader=csv.reader(file)
          for row in reader:
                print(row)
                
```
